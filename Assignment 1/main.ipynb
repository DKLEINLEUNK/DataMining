##### *Plots*

```python
entries_per_id = data.groupby('id')['variable'].size()

arguments = {
    'x_data': entries_per_id.index,
    'y_data': entries_per_id.values,
    'x_label': '',
    'y_label': 'Entries',
    'y_lim': (0, 27_500),
    'figsize': (10, 5),
    'color': 'skyblue',
    'rotate_x_labels': True,
    'show': True,
    'export': False,
    'add_values_to_bars': True,
    'add_values_to_bars_offset': 500,
    'title': None
}

descriptives.plot_barchart(**arguments)
```

```python
entries_per_variable = data['variable'].value_counts()

arguments = {
    'x_data': entries_per_variable.index,
    'y_data': entries_per_variable.values,
    'x_label': '',
    'y_label': 'Entries',
    'y_lim': (0, 120_000),
    'figsize': (10, 5),
    'color': 'skyblue',
    'rotate_x_labels': True,
    'show': True,
    'export': False,
    'add_values_to_bars': True,
    'add_values_to_bars_offset': 2_500,
    'title': None
}

descriptives.plot_barchart(**arguments)
```

```python
mood = data[data.variable == 'mood']
entries_per_mood = mood['value'].value_counts()

arguments = {
    'x_data': entries_per_mood.index,
    'y_data': entries_per_mood.values,
    'x_label': 'Mood',
    'y_label': 'Entries',
    'y_lim': (0, 3_000),
    'figsize': (10, 5),
    'color': 'lightcoral',
    'rotate_x_labels': False,
    'show': True,
    'export': False,
    'add_values_to_bars': True,
    'add_values_to_bars_offset': 75,
    'title': None
}

descriptives.plot_barchart(**arguments)
```
